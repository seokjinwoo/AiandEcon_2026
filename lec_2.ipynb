### 숫자(혹은 문자)를 담는 바구니들

- 파이썬에는 숫자를 담는 다양한 바구니들이 있음
- 리스트(list), 튜플(tuple), 배열(array), 데이터프레임(dataframe)

```python
# 리스트, bracket [ ]

# x = 1
# print(x)

x = [1,2,3,4,5]
print("x는", x)

y = [10, 9, 8]

x + y
```

- 파이썬 라이브러리를 사용하여 처리
- 수학, 숫자와 관련해서 가장 많이 쓰이는 라이브러리는 numpy
- numpy에 있는 array() 것을 이용해서 숫자들을 저장

```python
# numpy 로딩
import numpy as np

x = np.array([3,4,5])
# print(x)

y = np.array([4, 9, 7])

# x + y
x * y
```

```python
z = np.array([ [1,2,5], [3,4,8]  ])

print("z는", z.ndim, "차원입니다.")

print(z.dtype)
print(z.shape)
```

- 숫자를 리스트에 담는 작업.
- 어레이라고 하는 틀에다가 집어 넣었음




```python
# indexing 혹은 slicing 이라고도 함
# list 나 array에 넣은 숫자를 빼오는 작업

print(x)
print(x[0])
# [3,4]를 추출해보시오
```

```python
# 난수를 추출

x = np.random.normal(loc = 8, scale = 1, size = 50)
y = np.random.normal(loc = 8, scale = 1, size = 50)

# x 에 50개의 정규분포에서 추출된 숫자
# y 에 50개의 정규분포에서 추출된 숫자

# 두 확률변수 사이의 상관계수(correlation coefficient)
# -1 <=   <= 1 사이의 숫자.
# 1로 가까워지면, 양의 기울기를 갖는 직선
# -1로 가까워지면, 음의 기울기를 갖는 직선
# 0으로 가까워지면, 그런 관계가 없음

# print(np.corrcoef(x,y))
print(np.mean(x))
print(np.std(x))
```
