```python
import numpy as np
import pandas as pd

# 2개의 컬럼과 10개의 관측치를 표준정규분포에서 추출
data = np.random.randn(10, 2)

# 데이터프레임 생성
df = pd.DataFrame(data, columns=['X', 'Y'])

print(df)
```

```python
# 시각화(data visualization)
# X, Y로 구성된 산포도(scatter plot)

import matplotlib.pyplot as plt

# 산포도 그리기
plt.figure(figsize=(3,2))
plt.scatter(df['X'], df['Y'])
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot of Y against X')
plt.show()

```
